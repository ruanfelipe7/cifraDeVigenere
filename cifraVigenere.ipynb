```python
import numpy as np
import pandas as pd
```

### Leitura do arquivo com o texto encriptografado

```python
arquivo = open('out.enc', 'r')
textoCompleto = arquivo.read()
textoCompleto
```

### Retirando as pontuações para deixar apenas as palavras

```python
textoSemPontuacao = textoCompleto.replace(",", "")
textoSemPontuacao = textoSemPontuacao.replace("\n", "")
textoSemPontuacao = textoSemPontuacao.replace(".", "")
textoSemPontuacao
```

### Separando as palavras pelos espaços em branco

```python
palavras = np.array(textoSemPontuacao.split(" "))
palavras
```

### Selecionando apenas as palavras que contém apenas uma letra

```python
palavras1LetraCondicao = [len(x) == 1 for x in palavras]
palavras1Letra = palavras[palavras1LetraCondicao]
palavras1Letra
```

### Buscando os índices das palavras que possuem tamanho 1

```python
indices1letra = np.where([p1l == palavras for p1l in palavras1Letra])
indices = set(indices1letra[1])
ind = np.array(list(indices))
ind = np.sort(ind)
ind
```

### Combinando as palavras de tamanho 1 com seus respectivos indíces do vetor de palavras

```python
combinacao = []
for p1l, i1l in zip(palavras1Letra, ind):
    combinacao.append([p1l, i1l])
palavrasIndices = np.array(combinacao)
palavrasIndices
```

### Alterando os indíces das palavras de tamanho 1 para corresponder aos indíces no texto completo lido do arquivo

```python
tamAux = 0
for i in range(0, len(ind)):
    if i == 0:
        continue
    else:
        anterior = int(ind[i-1])
        j = int(palavrasIndices[i,1])
        while(j > anterior): 
            tamAux += len(palavras[j])
            j -= 1
        palavrasIndices[i, 1] = tamAux
        
palavrasIndices
```

### Função que testa entre as possibilidades de chaves quais são iguais dentro dos vetores senhas no campo i e o vetor caracteres no campo resto

```python
def testarIguais(senhas, i, caracteres, resto):
    senhas1 = sorted(senhas[i])
    senhas2 = sorted(caracteres[resto])
    senhasAux = []
    for sen1 in senhas1:
        for sen2 in senhas2:
            if sen1 == sen2:
                senhasAux.append(sen1)
    
    senhas[i] = senhasAux
    caracteres[resto] = senhasAux
```

### Função que encontra as possíveis chaves para as palavras de tamanho 1, orientando-se pelo fato de que no português apenas os caracteres 'o', 'a' e 'e' podem aparecer sozinhos em uma palavra

```python
def encontrarNovasSenhas(letra, senhas):
        vogaisLivres = [0, 4, 14]
        senhasDisponiveis = np.where(cifra == letra)
        senhasDisponiveis = senhasDisponiveis[1]
        senhasAux = [senhasDisponiveis[vogal] for vogal in vogaisLivres]
        senhas.append(senhasAux)

#p, l, b - Primeiro caractere        
```

### Encontra as possíveis chaves para cada uma das palavras de 1 caractere encontradas no arquivo e retorna essas chaves dentro da variável senhasDisp

```python
senhasDisp = []
for palInd in palavrasIndices:
    encontrarNovasSenhas(palInd[0], senhasDisp)
senhasDisp
```

### Função que busca entre as possibilidades para um tamanho máximo de chave, quais possíveis chaves podem ser utilizadas, retirando aquelas cuja correspondência com uma palavra de tamanho 1 não foi possível de ser realizada com as demais palavras de tamanho 1 que possuiriam a mesma chave no determinado caractere.

### O resultado das possíveis senhas do texto para cada caractere é retornado no array novasSenhas, que exclui aquelas senhas cujo não haviam nenhuma chave disponível para algum caractere.

```python
def buscarPossiveisSenhas(tamanhoMax, novasSenhas):
    for tam in range(1, tamanhoMax+1):
        senhasDisp = []
        for palInd in palavrasIndices:
            encontrarNovasSenhas(palInd[0], senhasDisp)
        print("Situação: ", tam)
        print("--------------------------")
        caracteres = []
        for j in range(0, tam):
            caracteres.append(-1)
        for ind in range(0, len(palavrasIndices)):
            resto = int(palavrasIndices[ind,1]) % tam
            if caracteres[resto] != -1:
                testarIguais(senhasDisp, ind, caracteres, resto)
            caracteres[resto] = senhasDisp[ind]
            print("Caractere cifrado:", palavrasIndices[ind,0], "---", caracteres)
        print("------------------------------")
        aux = []
        for c in caracteres:
            if len(c) == 0:
                break
            else:
                aux.append(c)
        if len(aux) == len(caracteres):
                novasSenhas.append(caracteres)
```

### Função que recebe um determinado caractere do texto e um caractere da chave e retorna o caractere descriptografado referente ao texto original

```python
def decript(caractere, chave):
    linha = np.where(cifra['0'] == caractere, cifra.columns, 0)
    result = linha[linha != 0]
    numCaractere = int(result)
    numChave = int(chave)
    line = ((numCaractere - numChave) + 26) % 26
    return cifra['0'][line]
```
